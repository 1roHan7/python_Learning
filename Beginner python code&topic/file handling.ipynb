# I_O processing 

```python
# I_O processing
with open("example.txt",'w') as file:
    content = file.write("Hello this is I_O processig \n")
    print(content)
```

```python

with open("example.txt",'r') as file:
    content = file.read()
    print(content)
```

```python
with open("example.txt",'a') as file:
    content = file.write("I'm appending this new line \n")
    print(content)
```

```python

with open("example.txt",'r') as file:
    for line in file :
        print (line.strip())
       
```

```python
import csv 
with open ('data.csv', 'r') as file:
    reader = csv.reader(file)
    data = [row for row in reader]
print (data)
```

```python
import csv 
with open ('data.csv', 'r') as file:
    reader = csv.DictReader(file)
    print(reader)
    for row in reader:
        print(row)
```

# Python Collection

```python
from collections import namedtuple,defaultdict,Counter
employee = namedtuple('Employee',['name','dept'])
e1 = employee("Rohan","123")
e2 = employee("bob","456")
print (e1,e2)
```

```python
d= defaultdict(list)
d['123'].append('Rohan')
print(d)
```

```python
c = Counter(['apple','banana','apple','apple'])
print(c['apple'])
print(c['banana'])
```

# Processing Python List

```python
data =[1,2,3,4,5]
sq = [_**2 for _ in data]
print(sq)
```

```python
evens = [x for x in data if x %2 ==0]
print(evens)
```

# Lambda function

```python
#lambda arguments : expression
square = lambda x: x*x
print(square(4))
```

```python
add = lambda a,b : a+b
print(add(5,6))
```

```python
# generally used with map () function 
# map (function,iterable)
data =[1,2,3,4,5]
squares = list(map(lambda x: x*x,data))
print(squares)
```

```python
# filter function
# map iterate over every items but filter will find the item equivalent to it 
evens = list(filter(lambda x : x%2 ==0,data))
odds = list(filter(lambda x : x%2 !=0,data))
print(evens)
print(odds)
```
