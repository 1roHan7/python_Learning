```python
data =[1,2,3,4,5]
sq = [_**2 for _ in data]
print(sq)
```

# Lambda function

```python
#lambda arguments : expression
square = lambda x: x*x
print(square(4))
```

```python
add = lambda a,b : a+b
print(add(5,6))
```

```python
# generally used with map () function 
# map (function,iterable)
data =[1,2,3,4,5]
squares = list(map(lambda x: x*x,data))
print(squares)
```

```python
# filter function
# map iterate over every items but filter will find the item equivalent to it 
evens = list(filter(lambda x : x%2 ==0,data))
odds = list(filter(lambda x : x%2 !=0,data))
print(evens)
print(odds)
```

# FILTERING DATA IN PYTHON

```python
data = [100,0,300,400]
filtered = list(filter(lambda x: x>150,data))
print(filtered)
```

```python
data =[100,-1,200,None,0]
clean_filtered = list(filter(lambda x:   x is not None and x>0, data))
print (clean_filtered)
```

```python
empoyees = [
    {'name': 'Rohan','salary':420000},
    {'name': 'SHalini','salary':410000},
    {'name': 'Bob','salary':120000}
]

high_earners = list(filter(lambda x: x['salary'] >400000, empoyees))
print(high_earners)
```

# Set() function 
- removing duplicates
- performing set operation like union, intersection, etc

```python
name = ['rohan' , 'shalini', 'kashvi', 'rohan', 'tannu']
unique_name = list(set(name))
print(unique_name)
```

```python
name = {'Shalini', 'tannu', 'rohan', 'kashvi'}
name1= {'shalini', 'rohan'}

non_common = name - name1
print(non_common)
```

```python
empoyees = [
    {'name': 'Rohan','salary':420000},
    {'name': 'SHalini','salary':420000},
    {'name': 'Bob','salary':120000}
]

uniques = set(map(lambda x: x['salary'],empoyees))
print(uniques)
```

# Sort Python list using Key

```python
# sorted(data,key)

people = [
    {'name':'alicess','age':'29'},
    {'name':'rohan bisht','age':'26'}
    ]
sorted_people = sorted(people,key =lambda x : x['age'])
print(sorted_people)

name = ['rohan', 'bob', 'charlie']
sorted_people1 = sorted(name,key = lambda x : len(x))
print(sorted_people1)
```

```python
def get_columns_name (schemas, ds_name, sorting_key = 'column_position'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key = lambda col : col[sorting_key])
    return[col['column_name'] for col in columns]
```
