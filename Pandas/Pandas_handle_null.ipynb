```python
import pandas as pd
```

```python
data = {
    'Name': ['Alice', 'Bob', None, 'David'],
    'Age': [25, None, 30, 22],
    'Score': [85, 90, None, 88]
}
df_main = pd.DataFrame(data)
display(df_main)
```

How to Use Dropna()

```python
#df.dropna() → drops rows with any nulls
df = df_main.dropna()
display(df)
```

```python
#df.dropna(how='all') → drops rows only if all values are null
df = df_main.dropna(how ='all')
display(df)
```

```python
#df.dropna(axis=1) → drops columns with any nulls
df = df_main.dropna(axis=1)
display(df)
```

```python
# df.dropna(subset=['Age', 'Score']) → drops rows where Age or Score is null
df = df_main.dropna(subset=['Age','Score'])
display(df)
```

Filling Nulls with fillna()

```python
# A constant (like 0 or "Unknown")
df = df_main.fillna('0')
display(df)
```

```python
# A computed value (like mean or median)
df= df_main.copy() #used to copy the df then update the df
df['Age'] = df['Age'].fillna(df['Age'].mean())
display(df)
```

```python
# Forward fill (propagate last valid value forward)
df = df_main.copy()
df = df.fillna(method='ffill')
display(df)
```

```python
# Backward fill (use next valid value)
df = df_main.copy()
df = df.fillna(method='bfill')
display(df)
```

```python
#How would you fill nulls in the 'Score' column with the median of that column?
df = df_main.copy()
df['Score'] = df['Score'].fillna(df['Score'].median())
display(df)
```

#Renaming Columns in Pandas

```python
df.rename(columns={'Score': 'Test_Score'}, inplace=True)
display(df)

#inplace=True tells pandas to apply the change directly to the original DataFrame — without creating or returning a new one.
```

```python
df = df_main.rename(columns={'Name' : 'Full_Name', 'Age' : 'Years'}, inplace=False)
display(df)
#inplace=true wont return anything while inplace =false will return a new dataframe with the changes
```

# Normalizing Data in Pandas

```python
# Normalize 'Years' column to range [0, 1]
df = df_main.copy()
df['Years_normalized'] = (df['Years'] - df['Years'].min()) / (df['Years'].max() - df['Years'].min())
display(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Score']] = scaler.fit_transform(df[['Score']])
display(df)
```
